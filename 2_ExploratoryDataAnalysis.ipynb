### 2. Spatial Distribution of N2 Fixation

```python
data = pd.read_excel("cleaned_data_2.xlsx")
```

```python
# plot y
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(data['Total N2 Fixation (μmol N m-2 d-1)'], bins=30, color='skyblue', edgecolor='black')
plt.title('Initial Distribution of Total N2 Fixation')
plt.xlabel('N2 fixation rate (μmol N m-2 d-1)')
plt.ylabel('Frequency')
plt.yscale('log')
plt.savefig('Initial Distribution of Total N2 Fixation.png')
plt.show()
```

```python
data.drop(data[data['Total N2 Fixation (μmol N m-2 d-1)'] > 4000].index, inplace=True)
data.describe()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Total N2 Fixation (μmol N m-2 d-1)'], bins=30, color='skyblue', edgecolor='black')
plt.title('Final Distribution of Total N2 Fixation')
plt.xlabel('N2 fixation rate (μmol N m-2 d-1)')
plt.ylabel('Frequency')
plt.yscale('log')
plt.savefig('Final Distribution of Total N2 Fixation.png')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from matplotlib.colors import LogNorm
```

```python
lat = data["LATITUDE"].values
lon = data["LONGITUDE"].values
fixation = data["Total N2 Fixation (μmol N m-2 d-1)"].values

fig, ax = plt.subplots(figsize=(12, 6), subplot_kw={"projection": ccrs.PlateCarree()})

ax.add_feature(cfeature.COASTLINE, linewidth=0.8)
#ax.add_feature(cfeature.BORDERS, linestyle=":")
ax.set_global()

sc = ax.scatter(
    lon, lat, c=fixation, cmap="viridis", 
    norm=LogNorm(vmin=1, vmax=10000),
    s=20, transform=ccrs.PlateCarree()
)

cbar = plt.colorbar(sc, orientation="vertical", fraction=0.03, pad=0.04)
cbar.set_label(r"Total N$_2$ Fixation (μmol N m$^{-2}$ d$^{-1}$)")

plt.title("Spatial Distribution of N$_2$ Fixation (Log Scale)")
plt.savefig('Spatial Distribution.png')
plt.show()
```

### 3. Distribution of Environmental Variables

```python
env_vars = ['MLD', 'CHL', 'SST', 'PAR', 'Salinity', 'DO', 'N_avg', 'P_avg', 'Fe_avg']

plt.figure(figsize=(15, 10))
for i, var in enumerate(env_vars, 1):
    plt.subplot(3, 3, i)
    plt.hist(data[var].dropna(), bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
    
    # Add log scale if needed (based on initial look)
    if data[var].skew() > 1:
        plt.yscale('log')
        #plt.title(f'Distribution of {var} (log scale)')
        
plt.tight_layout()
plt.savefig('env_vars.png')
plt.show()
```

#### Boxplots for Outlier Detection

### 5. Correlation Matrix

```python
import seaborn as sns
```

```python
df_renamed = data.rename(columns={'Total N2 Fixation (μmol N m-2 d-1)': 'N2 Fixation'})

select_cols = ['N2 Fixation', 'log_y', 'MLD', 'CHL', 'log_CHL', 'SST', 'PAR', \
               'Salinity', 'DO', 'N_avg', 'P_avg', 'Fe_avg', 'coord1', 'coord2', 'coord3', 'time_sin', 'time_cos']
corr_matrix = df_renamed[select_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', 
            center=0, linewidths=.5)
plt.title('Correlation Matrix')
plt.savefig('Correlation Matrix.png')
plt.show()
```

### 6. Split dataset

```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_excel('cleaned_data_2.xlsx')
```

```python
df.replace([float('inf'), float('-inf')], pd.NA, inplace=True)
# Drop all rows containing NaN values (which were originally inf or -inf)
df_cleaned = df.dropna()
# Optionally, save the cleaned dataframe
df_cleaned.to_excel('cleaned_data_2.xlsx', index=False)
```

```python
df = pd.read_excel('cleaned_data_2.xlsx')
```

```python

train_list = []
test_list = []

for month, group in df.groupby('month'):
    train_grp, test_grp = train_test_split(
        group,
        train_size=0.7,
        random_state=42,  # recur
        shuffle=True
    )
    train_list.append(train_grp)
    test_list.append(test_grp)

# merge
train_df = pd.concat(train_list).reset_index(drop=True)
test_df = pd.concat(test_list).reset_index(drop=True)

train_df.to_excel('train_data_by_month.xlsx', index=False)
test_df.to_excel('test_data_by_month.xlsx', index=False)
```

### Verify spatio features

```python
from sklearn.preprocessing import StandardScaler
env_vars = ['PAR', 'SST', 'log_CHL', 'MLD', 'Salinity', 'DO', 'N_avg', 'P_avg', 'Fe_avg']
df[env_vars] = StandardScaler().fit_transform(df[env_vars])
df
```

```python
df['grid_id'] = df['LATITUDE'].round().astype(int).astype(str) + "_" + df['LONGITUDE'].round().astype(int).astype(str)
df['group'] = df['grid_id'] + "_" + df['month'].astype(str)
```

```python
from sklearn.model_selection import GroupShuffleSplit

splitter = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
train_idx, test_idx = next(splitter.split(df, groups=df['group']))

train_df = df.iloc[train_idx].copy()
test_df = df.iloc[test_idx].copy()
```
